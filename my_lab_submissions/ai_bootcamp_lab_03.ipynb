# Task - 05

### Reshape matrix

```python
import numpy as np
temp, wind = df_fill_with_mean["temperature_C"].to_numpy(), df_fill_with_mean["wind_speed_kmph"].to_numpy() #changing this into numpy matrix
temp=temp.reshape(10,24)
```

```python
def find_mean(temp):
    return np.round(np.mean(temp, axis=1),2) # axis =1 for rows

def find_max(temp):
    return np.round(np.max(temp, axis=1),2)

def find_min(temp):
    return np.round(np.min(temp, axis=1),2)
```

```python
mean, mmax, min = find_mean(temp), find_max(temp), find_min(temp)
print(f"mean: {mean}, \nmax: {mmax}, \nmin: {mean}")
```

### Normalize

```python
def normalize(mat):
    return (mat-mat.mean())/np.std(mat)

normalize(temp)
```

### filter with mask

```python
mask = wind > 15
high_wind = wind[mask]
high_wind
```
